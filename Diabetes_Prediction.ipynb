**Importing the Depedencies**

**Logistic Regression**

```python
class Logistic_Regression():

  #self-> is the reference instant(object)
  #while creating a object we will give value of learning rate nd number of iteration
  # declaring learning rate & number of iterations (hyperparameters)
  def __init__(self, learning_rate, no_of_iterations):

    self.learning_rate = learning_rate
    self.no_of_iterations = no_of_iterations

  def fit(self, X, Y):#X=no.of features Y=training
  # number of data points & number of features
  # m-> no.of rows and n-> no.of cols

    self.m, self.n = X.shape #(769,8)
     # initiating the weights & bias
    self.w = np.zeros(self.n) # 8 cols
    self.b = 0
    self.X = X
    self.Y = Y

     #loop till no of iteration and change weight & bias value
     #gradient descent
    for i in range(self.no_of_iterations):
      self.update_weights(X,Y)
      #function to update weights & bias value
  def update_weights(self, X, Y):
       #y_hat formula(sigmoid fun)  Z = wX + b
       Y_hat = 1 / (1 + np.exp(-(self.X.dot(self.w) + self.b)))

        # derivatives    X=[769,8] Y=[769,1] (take transpose of X) then multiply(.)
       dw=(1/self.m)*np.dot(self.X.T , (Y_hat - self.Y))
       db=(1/self.m)*np.sum(Y_hat - self.Y)
       # updating the weights & bias using gradient descent
       self.w = self.w - self.learning_rate * dw
       self.b = self.b - self.learning_rate * db






  def predict(self, X):

    Y_pred = 1 / (1 + np.exp( - (X.dot(self.w) + self.b ) ))
    Y_pred = np.where( Y_pred > 0.5, 1, 0)
    return Y_pred
```

**Train Test Split**

```python
X_train,X_test,Y_train,Y_test=train_test_split(features,target,test_size=0.2,random_state=2)
```

**Now Training The Model**

```python
#call the Logistic model with input value
classifier=Logistic_Regression(learning_rate=0.01,no_of_iterations=1000)
# classifier=lr.Logistic_Regression(learning_rate=0.01,no_of_iterations=1000)
```

```python
#now fit the data
classifier.fit(X_train, Y_train)
```

**Model Evaluation**

**Accuracy Score**

```python
#accuracy score on training data
X_train_prediction=classifier.predict(X_train) #predicted by model
training_data_accuracy=accuracy_score(Y_train,X_train_prediction) #compare with test(Y_Train)
```
